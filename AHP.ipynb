```python
import numpy as np
import pandas as pd
```

```python
# Step 1: Load the dataset
import pickle
dataset = pd.read_pickle('dataset_final.pkl')
```

```python

# Step 2: Determine the pairwise comparison matrix and criteria weights

#[Portuguese, math , satifaction, Grades, Envolvement, Risk]

pairwise_matrix = [
    [1, 1, 7, 5, 7, 3],
    [1, 1, 7, 5, 7, 3],
    [1/7, 1/7, 1, 1/4, 1, 1/6],
    [1/5, 1/5, 4, 1, 4, 1/3],
    [1/7, 1/7, 1, 1/4, 1, 1/6],
    [1/3, 1/3, 6, 3, 6, 1]
]

pairwise_matrix

```

```python
column_sums = np.sum(pairwise_matrix, axis=0)

column_sums
```

```python
pairwise_matrix
```

```python
# Step 3: Calculate the normalized matrix 
#normalized_matrix = [np.divide(x,y) for x,y in zip(pairwise_matrix,column_sums)]
```

```python
j=0

normalized_matrix = [ [1, 1, 7, 5, 7, 3],
    [1, 1, 7, 5, 7, 3],
    [1/7, 1/7, 1, 1/4, 1, 1/6],
    [1/5, 1/5, 4, 1, 4, 1/3],
    [1/7, 1/7, 1, 1/4, 1, 1/6],
    [1/3, 1/3, 6, 3, 6, 1]
]


for i in range(0,len(normalized_matrix)):
    j = 0
    for k in column_sums:
        normalized_matrix[i][j]= np.divide(pairwise_matrix[i][j],k)    
        j=j+1
    i=i+1 
```

```python
column_sumsTot = np.sum(normalized_matrix, axis=0)

column_sumsTot
```

```python
weight_vector  = np.mean(normalized_matrix, axis=1)

weight_vector 
```

```python
row_mean_sum = np.sum(weight_vector)

row_mean_sum
```

```python
######### ANALISAR CONSISTENCIA
```

```python
pairwise_matrix
```

```python
criteria = np.multiply(weight_vector,pairwise_matrix )
criteria
```

```python
weight_sum_value = np.sum(criteria, axis=1)

weight_sum_value
```

```python
consistencia_col = np.divide(weight_sum_value,weight_vector )

consistencia_col
```

```python
lambdamax = np.mean(consistencia_col)

lambdamax
```

```python
consistencia = (lambdamax - 6) / 5

consistencia
```

```python
consistencia_ratio = consistencia / 1.24

consistencia_ratio
```

```python
dataset_pesos = np.multiply(dataset,weight_vector )
dataset_pesos
```

```python
coluna_ws = np.sum(dataset_pesos, axis=1)
coluna_ws
```

```python
ranked_data = dataset.copy()
ranked_data['Weighted Sum'] = np.sum(dataset_pesos, axis=1)
ranked_data = ranked_data.sort_values('Weighted Sum', ascending=True)
ranked_data['Rank'] = range(1, len(dataset) + 1)
```

```python
# Display the ranked data
for i, idx in enumerate(ranked_data.index):
    print(f"Rank {i+1}: AlunoId {idx+1}")

```
