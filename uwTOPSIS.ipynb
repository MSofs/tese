```python
import numpy as np
import pandas as pd
from uwTOPSIS.uwTOPSIS import *
```

```python
# Step 1: Load the dataset
import pickle
dataset = pd.read_pickle('dataset_final.pkl')
```

```python
features = dataset.values
features
```

```python
impacts = ["min", "min", "min", "min", "min","min"]
```

```python
#L: array with the lower bounds of the weigths.

L = np.repeat(0.1, dataset.shape[1])

L
```

```python
#U: array with the upper bounds of the weigths.
U = np.repeat(0.4, dataset.shape[1])

U
```

```python
norm = "euclidean" # "euclidean", "gauss", "minmax"
```

```python
#integer value for the L-p distance. (By default p=2)

p = 2
```

```python
x = uwTOPSIS(dataset, impacts, L, U, norm, p)
```

```python
x.keys()
```

```python
x['Ranking']['uwTOPSIS']
```

```python
ranked_data = dataset.copy()
ranked_data['Ranking'] = x['Ranking']['uwTOPSIS']
ranked_data = ranked_data.sort_values('Ranking', ascending=False)
ranked_data['Rank'] = range(1, len(dataset) + 1)

```

```python
# Display the ranked data
for i, idx in enumerate(ranked_data.index):
    print(f"Rank {i+1}: AlunoId {idx+1}")
```
