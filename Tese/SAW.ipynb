```python
l = ['Q20__nota_português_2.ºperiodo','Q26__nota_mat_2.ºperiodo','Q16_satisfação_com_notas_escolares','Q15_as_minhas_notas','envolvimento_scoretotal', 'Prev_Cluster']

for nome in l:
    for indice, valor in n_dataset[nome].iteritems():
        # Faça alguma operação desejada com o valor da linha
        # Por exemplo, imprima o índice e o valor da linha
        if(nome=='Q20__nota_português_2.ºperiodo' or nome =='Q26__nota_mat_2.ºperiodo' or nome== 'Q16_satisfação_com_notas_escolares' or nome=='Q15_as_minhas_notas' or nome=='Prev_Cluster') :
            if(valor==0.0): 
                n_dataset.loc[indice,nome ]= 0.15
            if(valor==1.0): 
                n_dataset.loc[indice, nome]= 0.30
            if(valor==2.0): 
                n_dataset.loc[indice, nome]= 0.45
            if(valor==3.0): 
                n_dataset.loc[indice, nome]= 0.60
            if(valor==4.0): 
                n_dataset.loc[indice, nome]= 0.75
            if(valor==5.0): 
                n_dataset.loc[indice, nome ]= 1


        if(nome=='envolvimento_scoretotal'):   
           
            if(valor>=0.0 and valor<=5.0): 
                n_dataset.loc[indice, 'envolvimento_scoretotal']= 0.15
            if(valor>5.0 and valor<=10.0): 
                n_dataset.loc[indice, 'envolvimento_scoretotal']= 0.30
            if(valor>10.0 and valor<=15.0): 
                n_dataset.loc[indice, 'envolvimento_scoretotal']= 0.45
            if(valor>15.0 and valor<=20.0): 
                n_dataset.loc[indice, 'envolvimento_scoretotal']= 0.60
            if(valor>20.0 and valor<25.0): 
                n_dataset.loc[indice, 'envolvimento_scoretotal']= 0.75
            if(valor>=25.0): 
                n_dataset.loc[indice, 'envolvimento_scoretotal']= 1

```

```python
normalized_data = n_dataset
```

```python
l = ['Q20__nota_português_2.ºperiodo','Q26__nota_mat_2.ºperiodo','Q16_satisfação_com_notas_escolares','Q15_as_minhas_notas','envolvimento_scoretotal', 'Prev_Cluster']

for nome in l:
    for indice, valor in n_dataset[nome].iteritems():
        # Faça alguma operação desejada com o valor da linha
        # Por exemplo, imprima o índice e o valor da linha
        min_value = n_dataset[nome].min()

        valor = n_dataset.loc[indice, nome]
        normalized_data.loc[indice, nome]= min_value / valor
        
```

```python
normalized_data
```

```python
# Perform the SAW calculation
weighted_data = normalized_data * weights

weighted_data
```

```python
# Calculate the weighted sum for each entry
weighted_sum = weighted_data.sum(axis=1)

weighted_sum
```

```python
# Rank the entries based on the weighted sum
ranked_data = dataset.copy()
ranked_data['Weighted Sum'] = weighted_sum
ranked_data = ranked_data.sort_values('Weighted Sum', ascending=False)
ranked_data['Rank'] = range(1, len(dataset) + 1)
```

```python
# Display the ranked data
for i, idx in enumerate(ranked_data.index):
    print(f"Rank {i+1}: Alternative {idx+1}")

```
