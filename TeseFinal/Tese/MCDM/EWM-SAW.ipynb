```python
import numpy as np
import pandas as pd
import math
```

```python
# Step 1: Load the dataset
import pickle
dataset = pd.read_pickle('dataset_final.pkl')
```

```python
dataset
```

```python
column_sum = np.sum(dataset,axis=0)

column_sum
```

```python
norm_values = np.divide(dataset,column_sum)

norm_values
```

```python
h = 1 / math.log(len(dataset))

h
```

```python
entropy = np.sum(norm_values * np.log2(norm_values))

entropy
```

```python
entropies = -h * entropy

entropies
```

```python
d = 1-entropies

d
```

```python
sum_d = np.sum(d)

sum_d
```

```python
weight_vector = np.divide(d,sum_d)

weight_vector
```

```python
weight_vector.tolist()
```

```python
dataset_pesos = np.multiply(dataset,weight_vector )
dataset_pesos
```

```python
ranked_data = dataset.copy()
ranked_data['Weighted Sum'] = np.sum(dataset_pesos, axis=1)
ranked_data = ranked_data.sort_values('Weighted Sum', ascending=True)
ranked_data['Rank'] = range(1, len(dataset) + 1)
```

```python
# Display the ranked data
for i, idx in enumerate(ranked_data.index):
    print(f"Rank {i+1}: AlunoId {idx+1}")

```
